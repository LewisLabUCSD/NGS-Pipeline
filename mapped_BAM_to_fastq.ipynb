# Extraction of mapped reads from bam files.

```Inputs: bam files generated by STAR or other aligners```

Navigate to the directory containging the BAM files.

If the BAM files do now follow this naming convention: ```'*.fastq.sort.bam'```, specifier your own file identifier in the second code bock

```Outputs: FASTQ files containing ONLY reads mapped to your reference genome (specified during reads alignment)```

````python
import os
picard_jar_loc = '/home/chihchung/Installation/picard.jar'
BAM_file_postfix = '.fastq.sort.bam'
````

````python
bamfs = [fs for fs in os.listdir('./') if fs.endswith(BAM_file_postfix) and not fs.startswith('mapped_')];bamfs
````

````python
import multiprocessing
import subprocess

def work(cmd):
    return subprocess.call(cmd, shell=True)
count = multiprocessing.cpu_count()
pool = multiprocessing.Pool(processes=count)
````

get mapped bam files

````python
if not os.path.isdir('mapped_fastq'): os.mkdir('mapped_fastq')
cmd_get_mapped = ['samtools view -b -F 4 %s > mapped_%s'% (bam_fn, bam_fn) for bam_fn in bamfs]
cmd_get_fastq_from_bams = [
'java -jar %s SamToFastq \
     I= mapped_%s \
     FASTQ= mapped_fastq/%s.fastq'%(picard_jar_loc, bam_fn, bam_fn[:-len(BAM_file_postfix)]) for bam_fn in bamfs]
````

````python
pool.map(work, cmd_get_mapped)
pool.map(work, cmd_get_fastq_from_bams)
````
